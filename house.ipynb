```python
bins = [0, 1000, 2000, 3000, 6000]     
group = ['Low','Average','High','Very high']
train['Area_bin'] = pd.cut(train['GrLivArea'], bins, labels=group)
Area_bin = pd.crosstab(train['Area_bin'], train['SalePrice'])
Area_bin= Area_bin.iloc[:, :20]
Area_bin.div(Area_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel('GrLivArea')
plt.ylabel('Percentage')
plt.show()
```

```python
bins = [0, 500, 1500, 6000]             
group = ['Low', 'Average', 'High']
train['BsmtArea_bin'] = pd.cut(train['TotalBsmtSF'], bins, labels=group)
BsmtArea_bin = pd.crosstab(train['BsmtArea_bin'], train['SalePrice'])
BsmtArea_bin= BsmtArea_bin.iloc[:, :20]
BsmtArea_bin.div(BsmtArea_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel('TotalBsmtSF')
plt.ylabel('Percentage')
plt.show()
```

```python
train['TotalArea'] = train['GrLivArea'] + train['TotalBsmtSF']
bins = [0, 1000, 2000, 3000, 10000]
group = ['Low', 'Average', 'High', 'Very high']
train['TotalArea_bin'] = pd.cut(train['TotalArea'], bins, labels=group)
TotalArea_bin = pd.crosstab(train['TotalArea_bin'], train['SalePrice'])
TotalArea_bin= TotalArea_bin.iloc[:, :20]
TotalArea_bin.div(TotalArea_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(5,4))
plt.xlabel('TotalArea (GrLivArea + Basement)')
plt.ylabel('Percentage')
plt.show()
```

```python
bins = [0, 5000, 10000, 50000]     
group = ['Low', 'Average', 'High']
train['LotArea_bin'] = pd.cut(train['LotArea'], bins, labels=group)
LotArea_bin = pd.crosstab(train['LotArea_bin'], train['SalePrice'])
LotArea_bin= LotArea_bin.iloc[:, :20]
LotArea_bin.div(LotArea_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(5,4))
plt.xlabel('LotArea')
plt.ylabel('Percentage')
plt.show()
```
